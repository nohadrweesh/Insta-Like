```python
def insta_like(image,insta_filter):
    retImge=None
    if insta_filter=="Nashville":
        retImge=nashville(image)
    elif insta_filter=="Lily":
        retImge=lilyFilter(image)
    elif insta_filter=="Proprocket":
        retImge=proproket(image)
    elif insta_filter=="LordKelvin":
        retImge=lordKelvin(image)
    elif insta_filter=="Lomo-fi":
        retImge=lomo_fi(image)
    else:
        print("this filter not implemented ")
    return retImge
        
```

## first filter
### adds an overlay (all pixelsare 255)then use gaussianblur filter

```python
def nashville(original_img):
    blurred=cv2.GaussianBlur(original_img,(3,3),0)
    size=original_img.shape#tuple
    overlayLayer=np.full(size,255,original_img.dtype)
    filttered_img=cv2.addWeighted(overlayLayer,0.3,blurred,0.7,0)
    blurred=cv2.GaussianBlur(filttered_img,(3,3),0)
    return blurred
```

## second filter
### adds an overlay (yellow overlay) then use gaussianblur filter

```python
def lilyFilter(original_img):
    size=original_img.shape#tuple
    blank_image = np.zeros(size,original_img.dtype)
    blank_image[:,:,0] = 255     
    blank_image[:,:,1] = 255
    blank_image[:,:,2] = 50
    filttered_img=cv2.addWeighted(blank_image,0.3,original_img,0.7,0)
    blurred=cv2.GaussianBlur(filttered_img,(3,3),0)
    return blurred
```

## third filter
### adds an overlay (red overlay) then use gaussianblur filter

```python
def proproket(orignal_img):
    size=orignal_img.shape#tuple
    blank_image = np.zeros(size,orignal_img.dtype)
    blank_image[:,:,0] = 240     
    blank_image[:,:,1] = 0
    blank_image[:,:,2] = 0
    filttered_img=cv2.addWeighted(blank_image,0.3,orignal_img,0.7,0)
    blurred=cv2.GaussianBlur(filttered_img,(3,3),0)
    
    
    return blurred
    
```

## fourth filter
### adds an overlay (brown overlay) then use gaussianblur filter

```python
def lordKelvin(original_img):
    size=original_img.shape#tuple
    blank_image = np.zeros(size,original_img.dtype)
    blank_image[:,:,0] = 255     
    blank_image[:,:,1] = 153
    blank_image[:,:,2] = 0
    filttered_img=cv2.addWeighted(blank_image,0.4,original_img,0.7,0)
    blurred=cv2.GaussianBlur(filttered_img,(3,3),0)
    return blurred
    
```
